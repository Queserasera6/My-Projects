```python
# Quadratic interpolation 

def quadratic(x_1, x_2, x_3, y_1, y_2, y_3):
    import numpy as np
    matrix_0 = np.array([[1, x_1, x_1**2], [1, x_2, x_2**2], [1, x_3, x_3**2]])
    matrix_1 = np.array([[y_1, x_1, x_1**2], [y_2, x_2, x_2**2], [y_3, x_3, x_3**2]])
    matrix_2 = np.array([[1, y_1, x_1**2], [1, y_2, x_2**2], [1, y_3, x_3**2]])
    matrix_3 = np.array([[1, x_1, y_1**2], [1, x_2, y_2**2], [1, x_3, y_3**2]])
    det_0 = np.linalg.det(matrix_0)
    det_1 = np.linalg.det(matrix_1)
    det_2 = np.linalg.det(matrix_2)
    det_3 = np.linalg.det(matrix_3)
    a = det_1/det_0
    b = det_2/det_0
    c = det_3/det_0
    
    return 'The equation is y = ',a,'x^2 +',b,'x +',c
    
    
```

```python
import numpy as np

# Define the function to solve the forward difference equation
def solve_forward_difference(a, b, h, y0, num_steps):
    y = [y0]  # Initialize the list to store the solution
    for i in range(1, num_steps + 1):
        y_next = y[-1] + h * (a * y[-1] + b)  # Forward difference formula
        y.append(y_next)
    return y

# Define the parameters of the forward difference equation
a = 0.1  # Coefficient a
b = 1.0  # Coefficient b
h = 0.1  # Step size
y0 = 1.0  # Initial condition
num_steps = 10  # Number of time steps

# Solve the forward difference equation
solution = solve_forward_difference(a, b, h, y0, num_steps)

# Print the solution
for i, y_i in enumerate(solution):
    print(f"Step {i}: y = {y_i:.4f}")
```
